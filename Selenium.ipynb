Q:[6] Write a python program to scrape data for Top 1000 Quotes of All Time.

```python
driver.get('http://www.azquotes.com/')
```

```python
tp=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
tp.click()
```

```python
Quotes=[]
Type=[]
Author=[]
```

```python
x=0
y=10
for p in range(x,y):
    quotes=driver.find_elements(By.XPATH,'//ul//div//a[@class="title"]')
    for i in quotes:
        Quotes.append(i.text)
    Next_=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]/a')
    Next_.click()
    time.sleep(5)
for q in range(x,y):
    type=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in type:
        Type.append(i.text)
    Next_=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]/a')
    Next_.click()
    time.sleep(5)
for s in range(x,y):
    author=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in author:
        Author.append(i.text)
    Next_=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]/a')
    Next_.click()
    time.sleep(5)
```

```python
len(Quotes),len(Author),len(Type)
```

Q:[7] Write a python program to display list of respected former Prime Ministers of India (i.e. Name,
Born-Dead, Term of office, Remarksf

```python
driver.get('https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1')
```

```python
table=[]
```

```python
data=driver.find_elements(By.XPATH,'//div[@class="TableData"]//tr')
for i in data:
    new=i.find_elements(By.XPATH,'./td')
    new_=[]
    for x in new:
        y=x.find_elements(By.XPATH,'./a')
        if y:
            new_.append(y[0].text)
        else:
            new_.append(x.text)
    table.append(new_)
```

```python
table
```

```python
slice=table[1:-1]
```

```python
slice_again=[[i[1],i[2],i[3] +'-'+ i[4],i[5]] for i in slice]
```

```python
columns=['Name','Age when assumed office','Terms of office','Duration']
```

```python
PM=pd.DataFrame(slice_again,columns=columns)
```
