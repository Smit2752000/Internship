4. Scrape the details of trending repositories on Github.com. 
Url = https://github.com/ 
You have to find the following details: 
A) Repository title 
B) Repository description 
C) Contributors count 
D) Language used

```python
driver.get('https://github.com/')
```

We are using ActionChains as we already used in above question.

```python
op=WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//li[@class="HeaderMenu-item position-relative flex-wrap flex-justify-between flex-items-center d-block d-lg-flex flex-lg-nowrap flex-lg-items-center js-details-container js-header-menu-item"][4]')))
Ac.move_to_element(op).perform()
trn=WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[4]/div/div[3]/ul/li[2]/a')))
trn.click()
```

```python
GitHub=[]
links=driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]//a')
lnk=[link.get_attribute('href') for link in links]
for i in lnk:
    try:
        driver.get(i)
        WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//p[@class="f4 my-3"]')))
        try:
            Desc=driver.find_element(By.XPATH,'//p[@class="f4 my-3"]').text
        except:
            Desc='-'
        try:
            Title=driver.find_element(By.XPATH,'//strong[@class="mr-2 flex-self-stretch"]').text
        except:
            Title='-'
        try:
            Count=driver.find_element(By.XPATH,'//span[@class="Counter ml-1"]').text
        except:
            Count='-'
        try:
            Lang=driver.find_element(By.XPATH,'//span[@class="color-fg-default text-bold mr-1"]').text
        except:
            Lang='-'
        GitHub.append({'Title':Title,'Description':Desc,'Contributours Count':Count,'Language':Lang})   
    except Exception as e:
        print(f'Error happen{e}')
    driver.back()
    time.sleep(3)
```

```python
Git=pd.DataFrame(GitHub)
```

7. Scrape the details most watched tv series of all time from imdb.com. 
Url = https://www.imdb.com/list/ls095964455/ You have 
to find the following details: 
A) Name 
B) Year span 
C) Genre 
D) Run time 
E) Ratings 
F) Votes 

```python
driver.get('https://www.imdb.com/chart/tvmeter/?ref_=nv_tvv_mptv')
```

```python
Tvshow=[]
```

```python
top=driver.find_elements(By.XPATH,'//div[@class="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-b189961a-9 bnSrml cli-title"]//a')
t_p=[i.get_attribute('href') for i in top]
for i in t_p:
    driver.get(i)
    WebDriverWait(driver,15).until(EC.presence_of_element_located((By.XPATH,'//span[@class="hero__primary-text"]')))
    try:
        name=driver.find_element(By.XPATH,'//span[@class="hero__primary-text"]').text
    except:
        name='-'
    try:
        yr=driver.find_element(By.XPATH,'//ul[@class="ipc-inline-list ipc-inline-list--show-dividers sc-ec65ba05-2 joVhBE baseAlt"]//li[2]').text
    except:
        yr='-'
    try:
        gnr=driver.find_element(By.XPATH,'//div[@class="ipc-chip-list__scroller"]').text
    except:
        gnr='-'
    try:
        run=driver.find_element(By.XPATH,'//ul[@class="ipc-inline-list ipc-inline-list--show-dividers sc-ec65ba05-2 joVhBE baseAlt"]//li[4]').text
    except:
        run='-'
    try:
        rate=driver.find_element(By.XPATH,'//span[@class="sc-c4ffe080-1 iQZtLP"]').text
    except:
        rate='-'
    try:
        vote=driver.find_element(By.XPATH,'//div[@class="sc-c4ffe080-3 hNEDIO"]').text
    except:
        vote='-'
    TV={'NAME':name,'GENRE':gnr,'YEAR':yr,'RUN_TIME':run,'RATING':rate,'VOTE':vote}
    Tvshow.append(TV)
    driver.back()
    time.sleep(3)
```

```python
TVSHOW=pd.DataFrame(Tvshow)
TVSHOW
```
